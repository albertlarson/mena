```python
import xarray as xr
import glob
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
x = sorted(glob.glob('v20_monthly_umm/*.nc'))
x1 = sorted(glob.glob('v21_monthly_umm/*.nc'))
len(x1)
```

```python
x1[0]
```

```python
times_v21 = np.arange(np.datetime64('2000-01'),
                      np.datetime64('2022-04'), np.timedelta64(1, 'M'))
```

```python
times = np.arange(np.datetime64('1970-01'),
                  np.datetime64('2015-01'), np.timedelta64(1, 'M'))
times.shape
```

```python
times[-1]
```

```python
xx = xr.open_dataset(x[0])
print(f"{xx.Rainfall.nbytes / 1000}\t\t\t# of kilobytes")
import time;
t0 = time.time();
print(np.nanmean(np.asarray(xx.Rainfall)));
t1=time.time();
print(f"{round(t1-t0,3)}\t\t\tseconds")

y = np.asarray(xx.Rainfall)
y.shape

plt.pcolormesh(y,cmap='cool')
plt.gca().set_aspect('equal')
```

```python
jjjj = []
for idx,i in enumerate(x):
    j = xr.open_dataset(i)
    jj = np.asarray(j.Rainfall)
    jjj = np.nanmean(jj)
    jjjj.append(jjj)
jjjj = np.asarray(jjjj)
print(f"{jjjj.shape}\t\t\tshape")
```

```python
kkkk = []
for idx,i in enumerate(x1):
    k = xr.open_dataset(i)
    kk = np.asarray(k.Rainfall)
    kkk = np.nanmean(kk)
    kkkk.append(kkk)
kkkk = np.asarray(kkkk)
print(f"{kkkk.shape}\t\t\tshape")
```

```python
fig = plt.figure()
gs = gridspec.GridSpec(2,1,height_ratios=[1,1],width_ratios=[1])

fig.add_subplot(gs[0])
plt.plot(times,jjjj)

fig.add_subplot(gs[1])
plt.plot(times_v21,kkkk)
```
